# Question 1 
## This algorithm uses trial division to test if the number input is a prime number

```python
import math
Val = int(input("What number would you like to check if it is prime?"))
if Val <=1:
    print(Val,"Is not a prime number")
else:
    prime = True
    for i in range(2, int(math.sqrt(Val)+1)):
        if Val % i == 0:
            prime = False
            break
    if prime == True:
        print(Val,"is a prime number")
    else:
        print(Val,"is not a prime number")
```

```python
import random
Guessed = False
totguess = 0
num = random.randrange(1,11)
while Guessed == False:
    Guess = int(input("Guess a number between 1 and 10"))
    if Guess > num and Guess <=10:
        print("Lower!")
        totguess += 1 
    elif Guess < num and Guess >=0:
        print("Higher!")
        totguess += 1
    elif Guess == num:
        totguess += 1
        print("Congratulations you guessed correctly!")
        print("It took",totguess,"Guesses")
        Guessed = True    
    else:
        print("Please input a number between 1 and 10")

    
```
