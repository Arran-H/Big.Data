```python
import math
Val = int(input("What number would you like to check if it is prime?"))
if Val <=1:
    print(Val,"Is not a prime number")
else:
    prime = True
    for i in range(2, int(math.sqrt(Val)+1)):
        if Val % i == 0:
            prime = False
            break
    if prime == True:
        print(Val,"is a prime number")
    else:
        print(Val,"is not a prime number")
```

```python
import math as m
x = float(input("input x"))
n = int(input("Input N"))

def equation(x,n):
    tot = 0.0
    for i in range(1,n + 1):
        tot += (x * (5 ** i) ** 2)/(i + m.factorial(i+1))
    return tot

result = equation(x,n)
print ("The result is: ",result)
```
